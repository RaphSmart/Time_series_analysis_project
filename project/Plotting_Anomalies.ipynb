## Plotting Anomalies from the complete dataset

```python
#Importing necessary libraries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df2 = pd.read_csv('g.csv') # Loading the dataset
#Necessary pre-processing
df2 = df2.sort_values(by='timestamp', ascending=True)
df2 = df2.replace({'label': {0.0: False, 1.0: True}})

df2.head(3)
df2['label'].value_counts()
```

```python
# Count of total samples in a dataset
Total_samples = max(df2['timestamp'].count(), df2['value'].count(), df2['label'].count())
print("Total samples are:", Total_samples)
# Count of normal data
normal_data_count = df2['label'].value_counts()[0]
print("Normal data count is:", normal_data_count)
# Count of anomaly data
anomaly_data_count = df2['label'].value_counts()[1]
print("Anomaly data count is:", anomaly_data_count)
```

```python
Anomaly_percentage = anomaly_data_count/Total_samples * 100
print("Total anomaly percentage in the dataset is:", Anomaly_percentage)
```

```python
# Plotting anomalies in the complete dataset
plt.figure(figsize=(16,4))
plt.title('Complete Dataset')
plt.scatter(df2.index, df2['value'], zorder=1, label = 'data')
plt.scatter(df2.index, df2['label']*df2['value'], color='r', marker='D', zorder=2, label = 'anomaly')
plt.ylabel('value', fontsize=12)
plt.xlabel('timestamp', fontsize=12)
plt.legend()
plt.show()
```
